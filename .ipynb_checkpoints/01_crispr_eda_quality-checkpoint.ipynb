```python
import os
from pyspark.sql import SparkSession

# Build Spark Session with Delta support
spark = SparkSession.builder \
    .appName("CRISPR_Data_Quality") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .config("spark.driver.extraJavaOptions", 
            "--add-opens=java.base/java.nio=ALL-UNNAMED " +
            "--add-opens=java.base/sun.nio.ch=ALL-UNNAMED") \
    .config("spark.executor.extraJavaOptions", 
            "--add-opens=java.base/java.nio=ALL-UNNAMED " +
            "--add-opens=java.base/sun.nio.ch=ALL-UNNAMED") \
    .getOrCreate()
spark.conf.set("spark.sql.execution.arrow.pyspark.enabled", "true")
# Load the Delta table we just created
# Using an absolute path is safer for network access
path = os.path.abspath("data/delta/gene_effects")
df = spark.read.format("delta").load(path)

# Show the top 10 genes by dependency score
df.sort("dependency_score", ascending=False).show(10)
```

```python
import os
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
import matplotlib.pyplot as plt
import seaborn as sns

# Reuse your existing session logic
spark = SparkSession.builder \
    .appName("CRISPR_Data_Quality") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()

df = spark.read.format("delta").load("data/delta/gene_effects")

# --- Step 1: Completeness Check ---
total_rows = df.count()
null_counts = df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns])

print(f"Total Records: {total_rows}")
null_counts.show()

# --- Step 2: Biological Sanity Check ---
# Are there specific genes that should always have low/high scores?
# TP53 is a common dependency in many lines. 
tp53_stats = df.filter(F.col("gene_symbol") == "TP53") \
               .select(F.mean("dependency_score"), F.stddev("dependency_score"))
tp53_stats.show()

# --- Step 3: Distribution Visualization ---
# Sample the data for plotting (Spark to Pandas for viz)
pdf = df.select("dependency_score").sample(0.01).toPandas()

plt.figure(figsize=(10, 6))
sns.histplot(pdf['dependency_score'], kde=True, bins=50)
plt.title("Distribution of CRISPR Dependency Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
